### [09] 함수로 만들고, 다른 지역의 데이터 수집 확인해보기

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd


def regional_news(sido, sigungu=None):
    # 크롬 드라이버 객체 생성 및 지정된 URL로 이동
    driver = webdriver.Chrome()
    driver.get('https://land.naver.com/news/region.naver') #네이버페이 부동산 - 우리동네뉴스
    
    # "지역선택" 요소 찾기
    xpath = '//*[@id="searchForm"]/dl/dt/span'
    element = driver.find_element(By.XPATH, xpath)
    
    # 시/도 드롭다운 요소를 찾아서 클릭
    city_list = '//*[@id="city_no"]'
    driver.find_element(By.XPATH, city_list).click()
    # 각 시/도별 요소 찾기
    # ➊ 시/도를 지정
    city = sido
    
    # 지정한 시/도에 해당하는 요소의 XPath를 동적으로 생성
    city_element = f'//select[@id="city_no"]/option[contains(text(), "{city}")]'
    
    # 해당 요소가 클릭 가능할 때까지 최대 10초간 대기
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, city_element)))
    driver.find_element(By.XPATH, city_element).click()  # 지정한 시/도의 요소를 찾아 클릭
    
    # ➋ 시/군/구 지정
    town = sigungu
    # 시/군/구 드롭다운 영역 다루기(시/군/구 값이 주어진 경우)
    if town:
     town_list = '//*[@id="dvsn_no"]'
     driver.find_element(By.XPATH, town_list).click()
     # 지정한 시/군/구에 해당하는 요소의 XPath를 동적으로 생성
     town_element = f'//*[@id="dvsn_no"]/option[contains(text(), "{town}")]'
     # 해당 요소가 클릭 가능할 때까지 최대 10초간 대기
     WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, town_element)))
     driver.find_element(By.XPATH, town_element).click() # 지정한 시/군/구 요소를 찾아 클릭
        
    # ➌ 1초간 추가 대기(드롭다운)
    time.sleep(1) 
    
    # 전체 뉴스 데이터 수집
    news = driver.find_elements(By.XPATH, '//*[@id="content"]/div/ul/li')
    
    rows = [] # 빈 리스트 생성
    
    
    for i, data in enumerate(news):
        # 뉴스 링크 추출
        link_element = data.find_element(By.XPATH, './dl/dt/a')
        link = link_element.get_attribute('href')
        
        # 제목과 내용 추출
        title, contents = data.text.split('\n', maxsplit = 1)  # maxsplit=1로 첫 번째 줄만 분리
        
        # row 생성
        row = {
            '뉴스기사명': title,
            '내용': contents,
            '링크': link
        }    
    
        # 이미지 URL 추출하기
        try:
            img = data.find_element(By.XPATH, './dl/dt/a/img').get_attribute('src')
            
            if img:  # 이미지 URL이 실제로 존재하는 경우만
                row['이미지'] = img
                
        except:
            row['이미지'] = None
                    
        rows.append(row)
            
    df = pd.DataFrame(rows) # 데이터프레임으로 변형    
    driver.quit() # 드라이버 종료
    
    # ➍ df를 반환
    return df
```
